```python
df=pd.read_csv('User_review.csv')
```

Rating Class : "Good" , "Average" ,"Bad"

```python
def class_r(rating):
  if rating>3:
    return 'Good'
  elif rating==3:
    return 'Average'
  else:
    return 'Bad'
  
df['Class']=df['overall'].apply(class_r)
df.to_csv('User_Review.csv',index=False)
```

```python
class_count=df['Class'].value_counts()
percent=class_count/len(df)*100
print(class_count,percent)
```

```python

df=pd.read_csv('User_review.csv')
gr=df[df['Class']=='Good']
ar=df[df['Class']=='Average']
br=df[df['Class']=='Bad']
gr=gr.sample(n=40500,random_state=1).sort_index()
updated_df=pd.concat([gr,ar,br])
updated_df.to_csv('User_review1.csv',index=False)
```

```python
del df
df=pd.read_csv('User_review1.csv')
class_count=df['Class'].value_counts()
percent=class_count/len(df)*100
print(class_count,percent)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

X=df['reviewText']
y=df['Class']
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(X)
print(tfidf_matrix)
```

```python
tfidf_matrix
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
idf_values = tfidf_vectorizer.idf_

feature_idf_dict = dict(zip(feature_names, idf_values))
for feature, idf in feature_idf_dict.items():
    print(f"Feature: {feature}, IDF: {idf}")
```

We have very skewed data.81% of the reviews is 'good'. 10% is 'bad' and only 8% is 'average'.
So, we need to modify out training dataset a bit.

Train and Test Data : 75:25

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

train_df=pd.DataFrame({'reviewText':X_train,'Class':y_train})
test_df=pd.DataFrame({'reviewText':X_test,'Class':y_test})

train_df.to_csv('train.csv',index=False)
test_df.to_csv('test.csv',index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier


train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
X_train = tfidf_vectorizer.fit_transform(train['reviewText'])
X_test = tfidf_vectorizer.transform(test['reviewText'])

y_train = train['Class']
y_test = test['Class']
```

```python
import gc
```

```python
#NAIVE BAYES
nb=MultinomialNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
with open('Naive_1.txt','w') as f:
  f.write("\nMultinomial Naive Bayes:\n")
  f.write(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred))
  
```

```python
del nb,y_pred
gc.collect()
```

```python
#Linear SVC
svc=LinearSVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
with open('SVC_1.txt','w') as f:
  f.write("\nLinear SVC:\n")
  f.write(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#Random Forest
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
with open('Random_Forest1.txt','w') as f:
  f.write("\nRandom Forest:\n")
  f.write(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred))  
  
```

```python
del rf,y_pred
gc.collect()
```

```python
#Gradient Boosting
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_pred=gb.predict(X_test)
with open('Gradient_1.txt','w') as f:
  f.write("\n Gradient Boosting:\n")
  f.write(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred))  
```

```python
del gb,y_pred
gc.collect()
```

```python
#MLP Classifier 
with open('evaluation5.txt','w') as f:
  mlp = MLPClassifier()
  mlp.fit(X_train, y_train)
  y_pred = mlp.predict(X_test)
  f.write("\nClassification Report for MLPClassifier:\n")
  f.write(classification_report(y_test, y_pred))
```
